# Part 1

Write a program to ask for your age as input, then output 1) whether or not you can vote and 2) whether your not you can retire. Let's assume the voting age is 18 or higher, and the retirement age is 65 or higher.

**NOTE:** This program is making two seprate decisions, and thus should have two separate if else statements.

Example Run:

```
Enter your age: 45
You can vote.
You cannot retire.
```


## Step 1: Problem Analysis

Inputs: Age

Outputs: Ability to Vote
         Ability to Retire

Algorithm (Steps in Program):
    1. Prompt for Age
    2. evaluate whether person can vote
    3. evaluate whether person can retire
    4. output if person can vote
    5. output if person can retire
    




````python
#Step 2: write code here
age = int(input ("Enter your age: "))
if age >= 18:
    print ("You can vote.")
else:
    print ("You can't vote.")
if age >= 65:
    print ("You can retire.")
else:
    print ("You can't retire.")
        
````

````python
## Step 2 (again): write code again but handle errors with try...except
try:
    age = int(input("Enter your age: "))
    if age <0:
        print ("That's not an age!")
    elif age <18 and age >0:
        print ("You can't vote. \nYou can't retire.")
    elif age >= 18 and age <65 :
        print ("You can vote. \nYou can't retire.")
    elif age >= 65:
        print ("You can vote. \nYou can retire.")
except ValueError:
        print ("That's not an age!")
````
