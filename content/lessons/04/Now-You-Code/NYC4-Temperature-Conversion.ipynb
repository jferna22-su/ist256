# Part 1

Write a Python program to input a temperature in Fahrenheit or Celcius,
then converts it to the other unit. The program should first input a temperature
then input a unit of C or F. It should then convert the temperature to the
other unit. 

After you get it working initially, re-write it to handle bad input. 

Example Run 1:

```
    Enter temperature: 212
    In which units is that: C or F? F
    That's 100.0 C
```

Example Run 2:

```
    Enter temperature: 0
    In which unit is that: C or F? C
    That's 32.0 F
```

NOTE: You can find the formulas online.


## Step 1: Problem Analysis

Inputs:
    - Temperature
    - What unit the temperature is

Outputs:
    -The temperature in the opposite unit

Algorithm (Steps in Program):
    1. prompt for temperature
    2. prompt for temperature unit
    3. caluclate other temperature unit
    4. output other temperature unit



````python
#Step 2: write code here
temperature = int(input("What's the temperature? "))
temp_unit = input("Is that in C or F? ")
if (temp_unit == 'c'):
    print ("That is %.1f F" % (temperature * 1.8 + 32))
elif (temp_unit == 'f'):
    print ("That is %.1f C" % ((temperature - 32) / 1.8 ))

````

````python
## Step 2 (again): write code again but handle errors with try...except
try:
    temperature = int(input("What's the temperature? "))
    temp_unit = input("Is that in C or F? ")
    if (temp_unit == 'c' or temp_unit == 'C'):
        print ("That is %.1f F" % (temperature * 1.8 + 32))
    elif (temp_unit == 'f' or temp_unit == 'F'):
        print ("That is %.1f C" % ((temperature - 32) / 1.8 ))
    else:
        print ("That's not a correct unit. You must enter C or F.")
except ValueError:
    print ("Sorry, that's not a temperature I can convert")
````
