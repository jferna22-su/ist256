````python
# Step 4: write main program here
from ipstack import GeoLookup
def geoiplookup2(ip):
    geo_lookup = GeoLookup("e8bb25e03c7b191185d25aca98b994b2")    
    state = geo_lookup.get_location(ip)
    return state

filename = "NYC2-IP-Addresses.txt"
with open(filename, 'r') as f:
    for line in f.readlines():
        location = geoiplookup2(line)
        print("IP:", location["ip"],"Location:", location["city"], ",", location ["region_code"])
````
