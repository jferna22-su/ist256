In this program you will implement an online shopping cart using a Python list of dictionary. The dictionary will contain the product name, price and quantity. 


The program should loop continually,  asking the user to enter 
 - Product name
 - Product price
 - Product quantity

until the user enters a product name of `'checkout'` at which time the loop should break.

Each time through the loop you should create a dictionary of product name, product price and product quantity then add the dictionary to a list.

After you enter `'checkout'`  the program should show:

 - all the items in the cart, including their quantity and price
 - and the total amount of the order, a running sum of quantity times price

## Step 1: Problem Analysis
```
write algorithm here
```

Inputs: User input of item name or checkout
        if user inputs item name, then the price of the item
        then user inputs quanity of item

Outputs:
        The item, its quanity and cost per unit
        The total cost of all items together

Algorithm (Steps in Program):
    1. create list for items
    2. create list for prices
    3. start loop
    4. create a dictionary for the items
    5. prompt the user for an item name or break statement
    6. if break statement is entered break the loop
    7. if item is entered prompt user for price of the item
    8. prompt the user for the quanity of the time
    9. calculate the total cost of the item with item cost times quanity
    10. add information to list for items
    11. add cost to list for prices
    12. once loop has been broken print out the item name, cost per item, and quanity as well as the total cost of everything   added together.

````python
# STEP 2: Write code
shopping_list = []
final_total = 0
while True:
    products = {'name' : '', 'price' : '', 'quanity' : '', 'total' : ''}
    user_products = input("Enter product name or 'checkout': ") 
    if user_products.upper() == "CHECKOUT":
        break
    else:
        products['name'] = (user_products.upper())
        products['price'] = float(input("What is the price of the %s? " % user_products ))
        products['quanity'] = int(input("How many %s do you need? " % user_products ))
        products['total'] = products['price'] * products['quanity']
        shopping_list.append(products)
        
for products in shopping_list:
    print (products['name'], products['quanity'],"$%.2f" % products['price'])
    final_total = final_total + products['total']
print('The total amount due is $%.2f' % final_total)
````
